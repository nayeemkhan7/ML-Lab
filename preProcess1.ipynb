```python
import pandas as pd
import numpy as np
import re
import string
import nltk
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Hates/labeled_data-MinV3.csv')
```

```python
df.head(10)
```

```python
#Remove pattern from the input text

def remove_pattern(input_text, pattern):
  r = re.findall(pattern, input_text)
  for word in r:
    input_text = re.sub(word, "", input_text)
  return input_text

```

```python
#Remove twitter handle (@username)

df['clean_tweet'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")
```

```python
df.head(10)
```

```python
#Remove special characters:

df['clean_tweet'] = df['clean_tweet'].str.replace("[^a-zA-Z#]", " ")
```

```python
df.head(10)
```

```python
#Tokenization

df['tokenized_tweet'] = df['clean_tweet'].apply(lambda x: x.split())
```

```python
df.head(10)
```

```python
#Stemming

from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

df['stemmed_tweet'] = df['tokenized_tweet'].apply(lambda sentence: [stemmer.stem(word) for word in sentence])
```

```python
df.head(10)
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
def remove_stopwords(tokenized_tweet):
  tweet_Xtop = [word for word in tokenized_tweet if word not in stopwords]
  return tweet_Xtop

df['xtop_tweet'] = df['tokenized_tweet'].apply(lambda x: remove_stopwords(x))
```
