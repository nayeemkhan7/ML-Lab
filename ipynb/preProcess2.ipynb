```python
import pandas as pd
import numpy as np
import re
import string
import nltk
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Hates/labeled_data-MinV3.csv')
```

```python
df.head()
```

```python
#Lowercasing

def lower_tweet(text):
  text = text.lower()
  return text
```

```python
df['lowercase_tweet'] = df['tweet'].apply(lower_tweet)
```

```python
df.head(10)
```

```python
#Remove pattern from the input text

def remove_pattern(input_text, pattern):
  r = re.findall(pattern, input_text)
  for word in r:
    input_text = re.sub(word, "", input_text)
  return input_text
```

```python
#Remove twitter handle (@username)

df['clean_tweet'] = np.vectorize(remove_pattern)(df['lowercase_tweet'], "@[\w]*")
```

```python
df.head(10)
```

```python
#Remove special characters:

df['clean_tweet'] = df['clean_tweet'].str.replace("[^a-zA-Z#]", " ")
```

```python
df.head(10)
```

```python
#Tokenization

tokenized_tweet = df['clean_tweet'].apply(lambda x: x.split())
tokenized_tweet.head(15)
```

```python
#Stemming

from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

stemmed_tweet = df['tokenized_tweet'].apply(lambda sentence: [stemmer.stem(word) for word in sentence])

stemmed_tweet.head(15)
```

```python
#Import stopwords list

nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
#Remove stopwords

def remove_stopwords(tokenized_tweet):
  tweet_Xtop = [word for word in tokenized_tweet if word not in stopwords]
  return tweet_Xtop

xtop_tweet = df['tokenized_tweet'].apply(lambda x: remove_stopwords(x))

xtop_tweet.head(15)
```

```python
#Combine again

for j in range(len(xtop_tweet)):
  xtop_tweet[j] = " ".join(xtop_tweet[j])

df['final'] = xtop_tweet
df.head(30)
```
